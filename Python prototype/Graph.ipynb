```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# ГРАФИК 1: Время выполнения запросов
# -----------------------------
data_sizes = [100000, 500000, 1000000]
rolap_times = [0.05, 0.09, 0.15]
oltp_times = [0.30, 0.90, 1.80]

plt.figure(figsize=(8,5))
plt.plot(data_sizes, rolap_times, marker='o', linewidth=2, label='ROLAP (ClickHouse)')
plt.plot(data_sizes, oltp_times, marker='s', linewidth=2, label='OLTP (PostgreSQL)')
plt.xlabel('Объём данных (число записей)')
plt.ylabel('Время выполнения запроса (сек)')
plt.title('График 1. Сравнение времени выполнения запросов')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('graph1_query_time.png', dpi=300)
plt.show()

# -----------------------------
# ГРАФИК 2: Зависимость времени от числа измерений
# -----------------------------
dimensions = [1, 2, 3, 4]
rolap_dim_times = [0.08, 0.12, 0.18, 0.25]
oltp_dim_times = [0.40, 0.95, 1.80, 3.10]

plt.figure(figsize=(8,5))
plt.plot(dimensions, rolap_dim_times, marker='o', linewidth=2, label='ROLAP (ClickHouse)')
plt.plot(dimensions, oltp_dim_times, marker='s', linewidth=2, label='OLTP (PostgreSQL)')
plt.xlabel('Количество измерений')
plt.ylabel('Время выполнения запроса (сек)')
plt.title('График 2. Влияние числа измерений на время выполнения')
plt.xticks(dimensions)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('graph2_dimensions.png', dpi=300)
plt.show()

# -----------------------------
# ГРАФИК 3: Влияние кэширования Redis
# -----------------------------
queries = [1, 2, 3, 4, 5]
without_cache = [0.25, 0.24, 0.26, 0.25, 0.24]
with_cache = [0.25, 0.08, 0.07, 0.07, 0.06]

plt.figure(figsize=(8,5))
plt.plot(queries, without_cache, marker='o', linewidth=2, label='Без кэша')
plt.plot(queries, with_cache, marker='s', linewidth=2, label='С кэшем Redis')
plt.xlabel('Номер запроса')
plt.ylabel('Время выполнения (сек)')
plt.title('График 3. Влияние кэширования Redis')
plt.xticks(queries)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('graph3_cache_effect.png', dpi=300)
plt.show()
```
